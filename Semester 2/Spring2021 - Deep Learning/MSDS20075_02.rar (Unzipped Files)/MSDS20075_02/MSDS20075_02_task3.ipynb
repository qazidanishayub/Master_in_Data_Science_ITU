```python
def loadDataset(path):
    print('Loading Dataset...')
    train_x, train_y, test_x, test_y = [], [], [], []
    for i in range(10):
        # print(i)
        for filename in glob.glob(path + '/ID3/' + str(i)+'/*.png'):
            # print(filename)
            im=plt.imread(filename)
            train_x.append(im)
            train_y.append(i)
    for i in range(10):
        for filename in glob.glob(path + '/testingdata/' + str(i)+'/*.png'):
            im=plt.imread(filename)
            test_x.append(im)
            test_y.append(i)
    print('Dataset loaded...')
    return np.array(train_x), np.array(train_y), np.array(test_x),np.array(test_y)
```

```python
train_set_x, train_set_y, test_set_x, test_set_y =  loadDataset('/content/drive/MyDrive/datasets/A3_train_bro/BroCheck')
```
