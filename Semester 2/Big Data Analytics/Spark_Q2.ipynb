```
dataRDD = sc.parallelize([("Brooke", 20), ("Denny", 31), ("Jules", 30), ("TD", 35), ("Brooke", 25), ("Denny", 102),
("TD", 39), ("Brooke", 40), ("Brooke", 81)
])
```

```
def sum_counts(a, b):
  return a + b
```

```
#apply tranformation
sum_RDD = dataRDD.reduceByKey(sum_counts)
```

```
sum_RDD.collect()
```

```
def avg_counts(a, b):
  return (a + b)/2
```

```
#apply tranformation
avg_RDD = dataRDD.reduceByKey(avg_counts)
```

```
avg_RDD.collect()
```

```
def squares(a):
  return (a*a)

def final(a):
  (a[2]/a[0] - (a[1]/a[0])**2)

def sum_values(a,b):
  (a[0]+b[0], a[1]+b[1], a[2]+b[2])
```

```
variance = dataRDD.mapValues(squares).reduceByKey(sum_values).mapValues(final)
```

```
variance.collect()
```
