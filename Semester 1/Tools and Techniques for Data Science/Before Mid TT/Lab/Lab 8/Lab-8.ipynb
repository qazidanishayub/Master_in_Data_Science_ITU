```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
#simply read the data and explore the data
df = pd.read_csv('FuelConsumption.csv')
df.describe()
df.shape
```

```
df.columns
```

```
#lets add some
cdf = df[['CYLINDERS','ENGINESIZE','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
cdf.head()
```

```
plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS, color='blue')
plt.xlabel('Engine Size')
plt.ylabel('Emission')
plt.show()
```

```
#splitting data into trainng and test data. 80% for traing and 20% for test
m = np.random.rand(len(cdf))<0.8
train = cdf[m]
test = cdf[~m]
```

```
c10 = int(input('enter first value c1: '))
c20 = int(input('enter second value c2: '))
gini_node0 = float( 1 - (c10/(c10+c20))**2 -(c20/(c10+c20))**2)
print('the gini  node 0 is : ',gini_node0)
t0 = c10 + c20
c11 = int(input('enter first value c1: '))
c21 = int(input('enter second value c2: '))
gini_node1 = float( 1 - (c11/(c11+c21))**2 -(c21/(c11+c21))**2)
print('the gini  node 0 is : ',gini_node1)
t1 = c11 + c21
c12 = int(input('enter first value c1: '))
c22 = int(input('enter second value c2: '))
gini_node2 = float( 1 - (c12/(c12+c22))**2 -(c22/(c12+c22))**2)
print('the gini is : ', gini_node2)
t2 = c12 + c22

combine = (t0/10 * gini_node0) + (t1/10 * gini_node1) + (t2/10 *gini_node2)
bro1 = float( 1 - ((c10+c11+c12)/10)**2 -((c20+c21+c22)/10)**2)
Gain_shareef = bro1 - combine 
print("bro shareef : ", bro1)
print("combine shareef : ", combine)
print('Gain_shareef : ', Gain_shareef)
```

```
#draw the tainning set
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS, color='red')
plt.xlabel('Engine Size')
plt.ylabel('Emission')
plt.show()
```

```
#SLR 
#Just follow the commands #builtin pakg 
from sklearn import linear_model
#made a object of linear regression reg
reg = linear_model.LinearRegression()
#my independdent variable
train_x = np.asanyarray(train[['CYLINDERS','ENGINESIZE']]) 
#my dependent variable y
train_y = np.asanyarray(train[['CO2EMISSIONS']])
reg.fit(train_x,train_y)
#the theta0 and theta1 print

print('Coefficients', reg.coef_)

print('Intercept', reg.intercept_)
```

```
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS, color='blue')
plt.plot(train_x,reg.coef_[0][0]*train_x + reg.intercept_[0],'r')
plt.xlabel('Engine Size')
plt.ylabel('CO2 Emisions')
```

```
new = [[4,2.0]]
reg.predict(new)
```

```
test_x = np.asanyarray(test[['ENGINESIZE']]) 
#my dependent variable y
test_y = np.asanyarray(test[['CO2EMISSIONS']])
y_hat = reg.predict(test_x)

print('Residual Error (MSE): %.2F'% np.mean((y_hat - test_y)**2))
print('Mean Absolute error: %.2F' %np.mean(np.absolute(y_hat - test_y)))
```

```
from sklearn.metrics import r2_score
print("R2 Score", r2_score(y_hat,test_y))
```
