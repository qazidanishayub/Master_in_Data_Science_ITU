```
df = pd.read_excel('canada.xlsx',sheet_name='Canada by Citizenship', skiprows=range(20),skipfooter=2)
```

```
df.head()
```

```
#rename columns
df.rename(columns={'OdName': 'Country', 'AreaName': 'Continent', 'RegName': 'Region' }, inplace=True)
```

```
df.columns
```

```
#column type into string
all(isinstance(column,str) for column in df.columns)
```

```
df.columns = df.columns.astype(str)
all(isinstance(column,str) for column in df.columns)
```

```
#Make Index Column for better selection
df.set_index('Country', inplace=True)
```

```
years = list(map(str,range(1980,2014)))
```

```
#task: plot a line graph of immigration from Haiti
haiti = df.loc['Haiti',years]
haiti.plot(kind='line')
plt.text(2000,6000,'2010 EarthQuake')
```

```
#lets compare china and pakistan from 1980 to 2013
df_PC = df.loc[['Pakistan', 'China'], years]

df_PC = df_PC.transpose()

df_PC.plot()


```

```
#compare the trends of top 5 countries that contributed towards canada
```

```
df['Total']= df.loc[:,years].sum(axis=1)
df['Total']
```

```
df_top5 = df.sort_values(by='Total', ascending=False).tail(5)
df_top5 = df_top5[years].transpose()
df_top5.plot(kind='area', stacked=False,figsize=(20,10))

```

```
#immigrants dist. of norway, denamrk and sweden in years

x = df.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
count, bin_edges = np.histogram(x,15)
print(bin_edges)
```

```
from wordcloud import WordCloud, STOPWORDS
```

```
hunger_game = open('hunger-game.txt','r').read()
stopwords = set(STOPWORDS)
hg = WordCloud(background_color='white',max_words=200,stopwords=stopwords)
#generate wordcloud
hg.generate(hunger_game)
#display
fig = plt.figure()
fig.set_figheight(15)
fig.set_figwidth(10)




plt.imshow(hg)
plt.axis('off')
plt.show

```
