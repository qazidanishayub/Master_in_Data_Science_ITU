```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img=cv2.imread("download.png")
print(img.shape)
plt.imshow(img)
plt.show()
```

```python
g=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
filter=np.array([1, -1])
# filter=filter.T
fx=cv2.filter2D(g,-1,filter)
fx[fx>0]=255
plt.imshow(fx, cmap='gray')
plt.show()
```

```python
img2=cv2.flip(img,-1)
plt.imshow(img2)
plt.show()
```

```python
img_rotated = cv2.rotate(img2, cv2.cv2.ROTATE_90_CLOCKWISE)
plt.imshow(img_rotated)
plt.show()
```

```python
img_blur=cv2.GaussianBlur(img_rotated, (3,3),0)
plt.imshow(img_blur)
plt.show()
```

```python
edges=cv2.Canny(img_blur, 50, 200)
plt.imshow(edges, cmap='binary')
plt.show()
```

```python
img_resized=cv2.resize(img_blur,(100,100))
plt.imshow(img_resized)
plt.show()
```

```python
gray=cv2.cvtColor(img_blur, cv2.COLOR_BGR2GRAY)
sobelx = cv2.Sobel(src=gray, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5)
plt.imshow(sobelx, cmap='gray')
plt.show()
```

```python
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])

img_prewittx = cv2.filter2D(gray, -1, kernelx)
img_prewitty = cv2.filter2D(gray, -1, kernely)
img_prewitt=img_prewittx+img_prewitty

plt.imshow(img_prewitt, cmap='gray')
plt.show()
```
